TODO B - Get a Pandas DataFrame for Daily Confirmed Cases in Nigeria. Columns are Date and Cases

Steps

* Create a new dataframe for plotting. This DataFrame will contain top 10 states in terms of confirmed cases i.e sort by confirmed cases. ** Hint: Check out Pandas [nlargest](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nlargest.html) function.
```python
#External Data 

df_ext_covid_data =  pd.read_csv(ext_covid_data)
df_ext_covid_data.head()
```

```python
# columns for ncdc data

df_ncdc_drop.columns
```

```python
#columns for external data

df_ext_covid_data.columns
```

```python
#merging the two dataset
df_merge_ext_ncdc = pd.merge(df_ncdc_drop, df_ext_covid_data, left_on = 'States', right_on = 'states')
df_merge_ext_ncdc.head()
```

```python
#sorted by confirmed cases

sorted_confirmed = df_merge_ext_ncdc.nlargest(10 , 'Confirmed cases' )
sorted_confirmed.tail()
```

```python
#Saving data in a DataFrame

df_sorted_confirmed_10= pd.DataFrame(sorted_confirmed)

df_sorted_confirmed_10
```

```python
#Plotting two variables on same axis

# create figure and axis objects with subplots()
fig,ax = plt.subplots()
plt.title (" Confirmed Cases and the overall community vulnerability index", fontsize=20)

# make a plot
lns1 = ax.plot(df_sorted_confirmed_10['States'], df_sorted_confirmed_10['Confirmed cases'], label='Confirmed cases', color="purple", marker="o")
# set x-axis label
ax.set_xlabel("States",fontsize=14)
# set y-axis label
ax.set_ylabel("Confirmed cases",color="purple",fontsize=14)



# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# making a plot for the 2nd y-axis using second axis object
lns2 = ax2.plot(df_sorted_confirmed_10['States'], df_sorted_confirmed_10['Overall CCVI Index'],label='Overall CCVI Index', color="green",marker="o")
ax2.set_ylabel("Overall CCVI Index",color="green",fontsize=15)

# Solution for having two legends
leg = lns1 + lns2
labs = [l.get_label() for l in leg]
plt.legend(leg, labs, loc=0)



plt.show()
```
